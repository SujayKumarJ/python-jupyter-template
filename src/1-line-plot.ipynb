### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt

x = list(range(11))
y = [i**2 for i in x]

plt.plot(x , y)

plt.title('Square of numbers from 0 to 10')
plt.xlabel('X values')
plt.ylabel('Y = xSquared')
plt.grid(True)

plt.show()
```
