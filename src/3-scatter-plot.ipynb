### Exercise 3: Scatter Plot with Annotations

```python
import matplotlib.pyplot as plt 

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.scatter(x,y)

for i in range(len(x)):
    plt.text(x[i], y[i], f'({x[i], y[i]})')

    plt.title('Scatter Plot with Annotations')

    plt.show()
```
